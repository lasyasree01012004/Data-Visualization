```python
import pandas as pd
import plotly.express as px
import folium
from folium import Marker, Popup, PolyLine
from IPython.display import display
```

```python
df = pd.read_csv('/content/listings.csv')

df = df.dropna(subset=['latitude', 'longitude'])
df = df.dropna(subset=['price'])

albany_lat = 42.6526
albany_lon = -73.7562

fig = px.scatter_mapbox(
    df,
    lat='latitude',
    lon='longitude',
    size='price',
    color='price',
    hover_name='name',  
    size_max=20,
    zoom=12,
)

fig.update_layout(mapbox_style="open-street-map")

fig.update_layout(
    geo=dict(
        center=dict(lat=albany_lat, lon=albany_lon),
        fitbounds="locations",
    ),
    title='Bubble Map of accomodation prices in Albany, New York, US',
)

fig.show()
```
